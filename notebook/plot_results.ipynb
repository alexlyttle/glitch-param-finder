```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = '/Volumes/daviesgr-alex-phd/gyraffe_jobs/results/m0.8.csv'
df = pd.read_csv(path)
```

```python
mask = (df['frac_age'] > 0.01) & (df['frac_age'] < 1.0)

ax = df.loc[mask].plot(x='nu_max', y='tau_he', c='star_feh', cmap='viridis', kind='scatter', alpha=0.2)
ax.set_xscale('log')
ax.set_yscale('log')
```

```python
ax = df.loc[mask].plot(x='nu_max', y='tau_cz', c='star_feh', cmap='viridis', kind='scatter', alpha=0.2)
ax.set_xscale('log')
ax.set_yscale('log')
```
